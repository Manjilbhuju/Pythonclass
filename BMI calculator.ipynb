```python
unit = int(input("Enter a unit "))
price = 0
if unit < 100:
    print("no charge".format(price))
elif unit <= 200:
    unit = unit - 100
    price = price + (unit * 5)
    print("{} is charged".format(price))
elif unit > 200 and unit <= 300:
    price = (100 * 5) + (unit - 200) * 10
    print("{} is charged".format(price))
else:
    unit > 300
    price = (100 * 5) + (200 * 10) + (unit - 300) * 15
    print("{} is charged".format(price))
```

```python
weight = int(input("Enter a number")) #KG
height = int(input("Enter a number")) #Cm
height = height/100
BMI=weight/height**2
if BMI <= 18.5:
    print("Underweight")
elif BMI >= 18.5 and BMI <= 24.9:
    print("Normal weight")
elif BMI >= 25 and BMI <= 29.9:
    print("Overweight")
else:
    print("Obesity")
```
